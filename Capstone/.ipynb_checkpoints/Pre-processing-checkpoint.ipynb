```python
# Lambda function for getting Athlete positions to append to Scouts dataframe.
def get_pos(row):
    id_ = df_atletas.loc[df_atletas.ID == row['AthleteID'], 'PosicaoID'].values[0]
    return id_
```

```python
'''
Function that crates a new feature called Away that is 1 when the player was playing away fro home and 0 for home turf games.
'''
def match_home_away(row):
    club_id = row['ClubID']
    match_round = row['Round']
    match_year = row['Season']
    res = df_partidas.loc[(df_partidas.CasaID == club_id) & (df_partidas.Rodada == match_round) & (df_partidas.Season == match_year), 'Resultado'].values
    if(len(res) > 0):
        return 0
    else:
        return 1
```

```python
# Creating Away column
df['Away'] = df.apply(match_home_away, axis=1)
```

```python
'''
Function that returns which adversary team was played against for each entry.
'''
def match_adversary(row):
    club_id = row['ClubID']
    match_round = row['Round']
    match_year = row['Season']
    away = row['Away']
    res = df_partidas.loc[((df_partidas.CasaID == club_id)|(df_partidas.VisitanteID == club_id)) & (df_partidas.Rodada == match_round) & (df_partidas.Season == match_year), ['CasaID', 'VisitanteID']].values
    if(len(res) > 0):
        if(away == 1):
            return res[0][0]
        else:
            return res[0][1]
    else:
        return 'NaN'
```

```python
# Adding opponentID column
df['opponentID'] = df.apply(match_adversary, axis=1)
```

```python
'''
Function that returns the team name for that round based on the Clubes spreadsheets.
This is for easier team identification.
'''
def team_name(row):
    club_id = row['ClubID']
    res = df_clubes.loc[(df_clubes.ID == club_id), 'Nome'].values
    if(len(res) > 0):
        return res[0]
    else:
        return 'NaN'
```

```python
# Team Names
df['Team'] = df.apply(team_name, axis=1)
```

```python
'''
Function that returns the opponent team name for that round based on the Clubes spreadsheets.
This is for easier opponent team identification.
'''
def team_name_opponent(row):
    club_id = row['opponentID']
    if(club_id != 'NaN'):
        res = df_clubes.loc[(df_clubes.ID == club_id), 'Nome'].values
        if(len(res) > 0):
            return res[0]
    else:
        return 'NaN'
```

```python
# Opponent Team Names
df['Opponent_Team'] = df.apply(team_name_opponent, axis=1)
```

```python
'''
Function that returns the athlete name for that round based on the Atletas spreadsheets.
This is for easier player identification.
'''
def athlete_name(row):
    athlete_id = row['AthleteID']
    if(athlete_id != 'NaN'):
        res = df_atletas.loc[(df_atletas.ID == athlete_id), 'Apelido'].values
        if(len(res) > 0):
            return res[0]
    else:
        return 'NaN'
```

```python
# Athlete Names
df['Athlete_Name'] = df.apply(athlete_name, axis=1)
```

```python
cols_scouts = ['Fouls_Received', 'Missed_Passes',
       'Assists', 'Shot_on_Post', 'Shot_Saved', 'Shot_Missed', 'Goals',
       'Offsides', 'Missed_Penalty', 'Stolen_Balls', 'Fouls_Conceded',
       'Own_Goals', 'Yellow_Card', 'Red_Card', 'Game_Without_Being_Scored',
       'Difficult_Saves', 'Penalty_Saves', 'Goals_Conceded']
```

```python
'''
Function that corrects the accumulation of all scouts for the corresponding season.
'''
def get_scouts_for_round(df, round_):
    suffixes = ('_curr', '_prev')
    cols_current = [col + suffixes[0] for col in cols_scouts]
    cols_prev = [col + suffixes[1] for col in cols_scouts]
    
    df_round = df[df['Round'] == round_]
    if round_ == 1: return df_round
    
    df_round_prev = df[df['Round'] < round_].groupby('AthleteID', as_index=False)[cols_scouts].max()
    df_players = df_round.merge(df_round_prev, how='left', on=['AthleteID'], suffixes=suffixes)
    
    # if it is the first round of a player, the scouts of previous rounds will be NaNs. Thus, set them to zero
    df_players.fillna(value=0, inplace=True)
    
    # compute the scouts 
    df_players[cols_current] = df_players[cols_current].values - df_players[cols_prev].values
    
    # update the columns
    df_players.drop(labels=cols_prev, axis=1, inplace=True)
    df_players = df_players.rename(columns=dict(zip(cols_current, cols_scouts)))
    df_players['Game_Without_Being_Scored']= df_players['Game_Without_Being_Scored'].clip_lower(0)
    
    return df_players
```

```python
df_scouts = df[df['Season'] != 2015]
df_scouts_2015 = df[df['Season'] == 2015]
n_rounds = df_scouts_2015['Round'].max()

if np.isnan(n_rounds):
    df_scouts = df
else:
    for i in range(1, n_rounds+1):
        df_round = get_scouts_for_round(df_scouts_2015, i)
        df_scouts = df_scouts.append(df_round, ignore_index=True)
    
print(df_scouts.shape)
df_scouts.head()
```

```python
df = df_scouts
```

```python
df_scouts = df[df['Season'] != 2017]
df_scouts_2017 = df[df['Season'] == 2017]
n_rounds = df_scouts_2017['Round'].max()

if np.isnan(n_rounds):
    df_scouts = df
else:
    for i in range(1, n_rounds+1):
        df_round = get_scouts_for_round(df_scouts_2017, i)
        df_scouts = df_scouts.append(df_round, ignore_index=True)
    
print(df_scouts.shape)
df_scouts.head()
```

```python
df = df_scouts
```
